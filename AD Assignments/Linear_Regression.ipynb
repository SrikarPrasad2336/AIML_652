```python
"""
Example: Cab Fare Calculation
Find Cab prediction price(y) by considering x value(No.of Km Traveled(20)) and rate per Km is fixed one(m) as 2 and fixed price as that is fare is c value is 5
Imagine a taxi company charges a base fare plus a cost per kilometer
Base fare (fixed charge): $5 (even if you travel 0 km)→ this is c.
Rate per km: $2 per km this is m.
Total fare: y depends on how many kilometers (x) you travel.
So, the equation for the taxi fare is
W342
compare the prices of ola and uber which one low price by
considering the below following things
Y=mx+c
fare price is 25 for ola
no.of km Traveled
ola: base price is 50
uber base price is 35
fare price:50"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

def calculate_fare(kilometers, rate_per_km, base_fare):
    total_fare = rate_per_km * kilometers + base_fare
    return total_fare

kilometers_traveled = 20
rate_per_km = 2
base_fare = 5

fixed_rate_fare = calculate_fare(kilometers_traveled, rate_per_km, base_fare)
print(f"Fixed Rate Fare for {kilometers_traveled} km: ${fixed_rate_fare}")

X = np.array([0, 5, 10, 15, 20, 25, 30]).reshape(-1, 1)
y = np.array([5, 15, 25, 35, 45, 55, 65])
model = LinearRegression()
model.fit(X, y)

distance = np.array([[20]])
predicted_fare = model.predict(distance)
print(f"Predicted Fare for 20 km using Linear Regression: ${predicted_fare[0]:.2f}")

plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('Kilometers Traveled')
plt.ylabel('Cab Fare ($)')
plt.title('Cab Fare Prediction using Linear Regression')
plt.legend()
plt.show()

ola_base_price = 50
ola_fare_per_km = 25
uber_base_price = 35
uber_fare_per_km = 50

ola_total_fare = calculate_fare(kilometers_traveled, ola_fare_per_km, ola_base_price)
uber_total_fare = calculate_fare(kilometers_traveled, uber_fare_per_km, uber_base_price)

print(f"Total Fare for Ola: ₹{ola_total_fare}")
print(f"Total Fare for Uber: ₹{uber_total_fare}")

if ola_total_fare < uber_total_fare:
    print("Ola offers a lower price.")
else:
    print("Uber is cheaper.")
```
